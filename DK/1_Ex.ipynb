```python
# suppose you have streaming messages on lat/lon, how could you determine which station is the closest for each message?
import json
from pyproj import Proj
from shapely.geometry import Point
from geopandas import GeoSeries, GeoDataFrame
```

```python
boros = GeoDataFrame.from_file('../data/nybb_13a/nybb.shp') # shp file downloaded from www.nyc.gov/html/dcp/html/bytes/applbyte.shtml
boros.head()
```

```python
boros.plot()
```

```python
gdf = GeoDataFrame(df)
s = GeoSeries([Point(x,y) for x, y in zip(gdf['lon'], gdf['lat'])])
gdf['geometry'] = s
gdf.crs = {'init': 'epsg:4326', 'no_defs':True}
gdf.geometry.total_bounds
gdf.head()
```

```python
gdf.to_crs(boros.crs, inplace=True)
gdf.geometry.total_bounds
```

```python
manhattan = boros.geometry[3]
brooklyn = boros.geometry[2]
in_mn = gdf.geometry.within(manhattan)
in_bk = gdf.geometry.within(brooklyn)

print sum(in_mn), 'stations in Manhattan'
print sum(in_bk), 'stations in Brooklyn'

print sum(gdf['bikes_avail'][in_mn]), 'available bikes in Manhattan'
print sum(gdf['bikes_avail'][in_bk]), 'available bikes in Brooklyn'
```
