__Regression__

```python
from sklearn import linear_model

lr = linear_model.LinearRegression()

# get data
carat = [[x] for x in gg.diamonds['carat'].values]
px = gg.diamonds['price']

#carat vs px
fit1 = lr.fit(carat, px)
```

```python
print -fit1.intercept_/fit1.coef_
```

```python
lr_zero = linear_model.LinearRegression(fit_intercept=False)
fit2 = lr_zero.fit(carat, px)

r2 = lr_zero.score(carat, px)
```

__Multivariate Regression__

```python
from sklearn import preprocessing

def convert_label(data):
    le = preprocessing.LabelEncoder()
    # fit label encoder
    le.fit(data)
    # print classes
    print list(le.classes_)
    # transform
    out = le.transform(data)
    return(out)
```

```python
gg.diamonds['colorFactor'] = convert_label(gg.diamonds.color)
```
